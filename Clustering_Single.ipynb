```python
#Import
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Array point
dp=np.array([
    [3, 7],
    [2, 6],
    [2, 2],
    [5, 8],
    [5, 5],
    [6, 6],
    [8, 4],
    [5, 2],
    [7, 3],
    [12, 8],
    [10, 6]
])
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram

#Single_Link
single_linkage_matrix=linkage(dp,method='single')

#Average_Link
average_linkage_matrix=linkage(dp,method='average')

#Complete_Link
complete_linkage_matrix=linkage(dp,method='complete')

#visualizedendrograms
fig,ax=plt.subplots(3,1,figsize=(9,10))

#Single_Link Dendrogram

dendrogram(single_linkage_matrix, ax=ax[0], labels=[str(point) for point in dp])
ax[0].set_title('Single Link Hierarchical Clustering')
ax[0].set_xlabel('Array Points')
ax[0].set_ylabel('Distance')

#Average_Link Dendrogram
dendrogram(average_linkage_matrix, ax=ax[1], labels=[str(point) for point in dp])
ax[1].set_title('Average Link Hierarchical Clustering')
ax[1].set_xlabel('Array Points')
ax[1].set_ylabel('Distance')

#Complete_Link Dendrogram
dendrogram(complete_linkage_matrix, ax=ax[2], labels=[str(point) for point in dp])
ax[2].set_title('Complete Link Hierarchical Clustering')
ax[2].set_xlabel('Array Points')
ax[2].set_ylabel('Distance')

plt.tight_layout()
plt.show()
```
